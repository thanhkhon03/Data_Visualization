```python
#Bài tập:Hãy tạo ra một line chart để biểu diễn sự thay đổi 
#của nhiệt độ trong một khoảng thời gian nhất định
import matplotlib.pyplot as plt
import numpy as np
days_of_week = ['Thứ 2','Thứ 3','Thứ 4','Thứ 5','Thứ 6','Thứ 7','Chủ nhật']
temperature = [25,27,28,26,29,30,28]

plt.plot (days_of_week, temperature,'x:')
plt.title (' Biểu đồ sự thay đổi của nhiệt độ trong một tuần')
plt.xlabel ('Ngày')
plt.ylabel ('Nhiệt độ')

# Thêm lưới
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
cua_hang = ['Cửa hàng A','Cửa hàng B','Cửa hàng C','Cửa hàng D','Cửa hàng E',]
so_luong = [150,180,200,220,190]

plt.bar (cua_hang, so_luong,color = 'aqua')
plt.title ('So sánh số lượng sản phẩm bán được giữa các cửa hàng')
plt.xlabel('Cửa hàng')
plt.ylabel('Số lượng sản phẩm bán được')
plt.grid(color='gray', linestyle=':',axis='y') #Tạo lưới 
plt.xticks (rotation = 45)
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['Áo', 'Quần', 'Giày', 'Túi xách', 'Đồng hồ']
sales = [250,180,150,120,90]

explode = [0.0,0.0,0.0,0.0,0.08]
plt.pie(sales, labels = products,explode = explode, autopct = '%.2f%%', startangle=140)

plt.title('Phân phối các loại sản phẩm bán được')

plt.show()
```
