# Yêu cầu 2

```python
# Đọc dữ liệu từ GitHub vào DataFrame
df = pd.read_csv(url)

# Vẽ biểu đồ đường
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Daily Confirmed'], label='Daily Confirmed', color='blue')

# Đặt tiêu đề và nhãn cho biểu đồ
plt.title('Số ca covid hàng ngày ở Ấn Độ')
plt.xlabel('Ngày')
plt.ylabel('Số ca ghi nhân')
plt.xticks(rotation=45)
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500])
plt.legend()


```

```python
skiprows = list(range(1, 62))  # Skip các dòng từ 1 đến 62
take_rows = list(range(62, 107))  # Chỉ lấy các dòng từ 63 đến 106
df = pd.read_csv(url, skiprows=skiprows, nrows=len(take_rows))
# Vẽ biểu đồ đường
plt.figure(figsize=(20,5))
plt.plot(df['Date'], df['Daily Confirmed'],'o-', label='Daily Confirmed', color='aqua')

# Đặt tiêu đề và nhãn cho biểu đồ
plt.gca().set_facecolor('black')
plt.title('COVID 19 IN INDIA')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.xticks(rotation=45)
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500])
plt.grid(color='lightyellow',linestyle='-',axis='both')
```

# Yêu cầu 4

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://github.com/nv-thang/Data-Visualization-Course/raw/main/covid19.csv"

skiprows = list(range(1, 62))  # Skip các dòng từ 1 đến 62
take_rows = list(range(62, 107))  # Chỉ lấy các dòng từ 63 đến 106
df = pd.read_csv(url, skiprows=skiprows, nrows=len(take_rows))

# Vẽ biểu đồ đường
plt.figure(figsize=(20,5))
plt.plot(df['Date'], df['Daily Confirmed'],'o-', label='Daily Confirmed', color='aqua')
plt.plot(df['Date'],df['Daily Recovered'],'o-',label='Daily Recovered',color='lime')
plt.plot(df['Date'],df['Daily Deceased'],'o-',label='Daily Deceased',color='red')


# Đặt tiêu đề và nhãn cho biểu đồ
plt.gca().set_facecolor('black')
plt.title('COVID 19 IN INDIA')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.xticks(rotation=45)
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500])
plt.grid(color='lightyellow',linestyle='-',axis='both')
plt.legend(['Số ca được xác nhận','Số ca được phục hồi','Số ca tử vong'])
```
